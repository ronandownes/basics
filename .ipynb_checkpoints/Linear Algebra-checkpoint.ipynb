```python
import matplotlib.pyplot as plt

scalar = 5

plt.plot(scalar, 0, 'ro')
plt.xlim(-10, 10)
plt.ylim(-1, 1)
plt.axhline(0, color='black',linewidth=0.5)
plt.xlabel('Magnitude')
plt.title('Geometric Perspective of a Scalar')
plt.show()
```
