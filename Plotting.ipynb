```python
import matplotlib.pyplot as plt

years = [1900, 1910, 1920, 1930, 1940, 1950, 1960, 1970, 1980, 1990, 2000, 2010]

uk_male_life_expectancy = [46.6, 51.3, 53.8, 58.4, 63.2, 66.0, 68.9, 70.4, 72.0, 74.1, 75.8, 78.0]
uk_female_life_expectancy = [50.2, 55.2, 58.3, 62.7, 66.9, 70.2, 73.5, 76.0, 78.3, 79.9, 80.9, 82.1]

us_male_life_expectancy = [46.3, 48.4, 53.6, 58.1, 61.6, 65.6, 66.6, 67.1, 70.0, 71.8, 74.3, 76.2]
us_female_life_expectancy = [48.3, 51.8, 54.6, 61.6, 65.7, 71.1, 73.1, 74.7, 77.4, 78.8, 79.7, 80.8]

plt.plot(years, uk_male_life_expectancy, label='UK Male')
plt.plot(years, uk_female_life_expectancy, label='UK Female')
# plt.plot(years, us_male_life_expectancy, label='US Male')
# plt.plot(years, us_female_life_expectancy, label='US Female')

plt.xlabel('Year')
plt.ylabel('Life Expectancy')
plt.title('Historical Life Expectancy: UK vs US')
plt.legend()

plt.show()
```
